```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Quarter": ["Q1", "Q2", "Q3", "Q4"],
    "Revenue ($B)": [28.6, 32.0, 34.1, 40.1]  # replace with actual
}

df = pd.DataFrame(data)
```

### Bar By Quarterly Revenue

```python
plt.figure(figsize=(8,5))
plt.bar(df["Quarter"], df["Revenue ($B)"])

plt.title("Meta Quarterly Revenue")
plt.xlabel("Quarter")
plt.ylabel("Revenue (Billions USD)")

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Quarter"], df["Revenue ($B)"], marker='o')

plt.title("Meta Revenue Trend")
plt.xlabel("Quarter")
plt.ylabel("Revenue (Billions USD)")

plt.show()
```

### Revenue By Numbers

```python
data = {
    "Year": ["2022", "2023", "2024"],
    "Revenue ($B)": [116.6, 134.9, 164.5]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Revenue ($B)"], marker='o')

plt.title("Meta Revenue Growth (2022–2024)")
plt.xlabel("Year")
plt.ylabel("Revenue (Billions USD)")

for i, value in enumerate(df["Revenue ($B)"]):
    plt.text(i, value, str(value), ha='center')

plt.show()
```

### Meta Revenue (2022-24)

```python
data = {
    "Year": ["2022", "2023", "2024"],
    "Revenue (USD Billions)": [116.6, 134.9, 164.5]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.bar(df["Year"], df["Revenue (USD Billions)"])

plt.title("Meta Revenue (2022–2024)")
plt.xlabel("Year")
plt.ylabel("Revenue (Billions USD)")

plt.show()
```

### The Revenue Heatmap

```python
data = {
    "2022": [116.6],
    "2023": [134.9],
    "2024": [164.5]
}

df = pd.DataFrame(data, index=["Revenue"])

plt.figure(figsize=(6,2))
sns.heatmap(df, annot=True, fmt=".1f", cmap="Blues")

plt.title("Meta Revenue Heatmap (USD Billions)")
plt.show()
```

### Cash Flow Vs The Revenue

```python
data = {
    "Year": ["2022", "2023", "2024"],
    "Revenue": [116.6, 134.9, 164.5],
    "Cash Flow": [50.5, 71.0, 91.3]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Revenue"], marker='o', label="Revenue")
plt.plot(df["Year"], df["Cash Flow"], marker='o', label="Operating Cash Flow")

plt.title("Meta Revenue vs Cash Flow (2022–2024)")
plt.xlabel("Year")
plt.ylabel("Billions USD")
plt.legend()

plt.show()
```

### CASH FLOW

```python
# Chart 2: Operating Cash Flow
plt.figure(figsize=(8,5))
plt.bar(df["Year"].astype(str), df["Cash Flow"])
plt.title("Meta Operating Cash Flow (2023–2025)")
plt.xlabel("Year")
plt.ylabel("Operating Cash Flow (Billions USD)")

for i, v in enumerate(df["Cash Flow"]):
    plt.text(i, v + 1, f"{v:.1f}", ha="center")

plt.show()
```
